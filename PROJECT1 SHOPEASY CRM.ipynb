```python
# Create Table query in Sql
## python to sql

tab1 = """

create table IF NOT EXISTS Geography
(
    GeographyID int PRIMARY KEY,
    Country varchar(20),
    City varchar(20)
);

"""
tab2 = """

create table IF NOT EXISTS Customers
(
    CustomerID int PRIMARY KEY,
    CustomerName varchar(50),
    Email varchar(50),
    Gender varchar(10),
    Age int,
    GeographyID int,

    FOREIGN KEY (GeographyID) references Geography(GeographyID)
);

"""
tab3 = """

create table IF NOT EXISTS Products
(
    ProductID int PRIMARY KEY,
    ProductName varchar(50),
    Category varchar(50),
    Price float
);


"""
tab4 = """

create table IF NOT EXISTS Engagement_data
(
    EngagementID int PRIMARY KEY,
    ContentID int  NOT NULL,
    ContentType varchar(50),
    Likes int,
    EngagementDate Date,
    CampaignID int NOT NULL,
    ProductID int ,
    ViewsClicksCombined varchar(50),
    Views int,
    Clicks int,

    FOREIGN KEY (ProductID) references Products(ProductID)
);
"""
tab5 = """

create table IF NOT EXISTS  customer_journey
(
    JourneyID int PRIMARY KEY,
    CustomerID int ,
    ProductID int ,
    VisitDate Date,
    Stage varchar(50),
    Action varchar(50),
    Duration  int,
    

    FOREIGN KEY (ProductID) references Products(ProductID),
    FOREIGN KEY (CustomerID) references Customers(CustomerID)
);


"""

tab6 = """

create table IF NOT EXISTS customer_reviews
(
    ReviewID int PRIMARY KEY,
    CustomerID int ,
    ProductID int ,
    ReviewDate Date,
    Rating int,
    ReviewText text(500),    

    FOREIGN KEY (ProductID) references Products(ProductID),
    FOREIGN KEY (CustomerID) references Customers(CustomerID)
);


"""

```

```python
# Excute The Create Table Query 
curr.execute(tab1)
curr.execute(tab2)
curr.execute(tab3)
curr.execute(tab4)
curr.execute(tab5)
curr.execute(tab6)
```

```python
# Sentiment Analysis for Reviews  - Query -5

sen_ana = """ SELECT 
                    cr.ReviewText, 
                    CASE 
                        -- ðŸ”¹ First check for Neutral condition (this must be checked first)
                        WHEN LOWER(TRIM(cr.ReviewText)) REGEXP 'but[[:space:]]+could[[:space:]]+be[[:space:]]+cheaper' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%good quality, but could be cheaper%' 
                        THEN 'Neutral' 
                        
                        -- ðŸ”¸ Negative Sentiments
                        WHEN REPLACE(cr.ReviewText, '  ', ' ') LIKE '%Not worth the money%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%bad%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%poor%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%terrible%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%disappointed%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Not worth%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Money%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Late%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Stopped%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%did not%'
                             
                        THEN 'Negative'
                
                        -- ðŸ”¹ Positive Sentiments (checked last)
                       WHEN REPLACE(cr.ReviewText, '  ', ' ') LIKE '%Five stars for the quick delivery%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%good%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%excellent%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%satisfied%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%amazing%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%great%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%top-notch%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Exceeded%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '% buy again!%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%love%'                             
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '% quick delivery%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%impressive%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Five stars%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%fast%'
                              
                        THEN 'Positive'
                        
                        -- Default to Neutral
                        ELSE 'Neutral'  
                    END AS Sentiment
                FROM customer_reviews cr;
                
                """
```

```python
# Negative Reviews Affect On Products
Neg_Prod = """SELECT p.ProductName,cr.ReviewText,
                     COUNT(cr.ReviewText) AS NegativeReviewCount
                     FROM customer_reviews cr
                     JOIN products p ON cr.ProductID = p.ProductID 
                     Where REPLACE(cr.ReviewText, '  ', ' ') LIKE '%Not worth the money%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%bad%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%poor%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%terrible%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%disappointed%' 
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Not worth%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Money%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Late%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%Stopped%'
                             OR LOWER(TRIM(cr.ReviewText)) LIKE '%did not%'                             
                        
                     GROUP BY  p.ProductName,cr.ReviewText
                     ORDER BY NegativeReviewCount DESC;
"""
```

```python
# Rating PRODUCTS - Query - 07
rating ="""SELECT Distinct(p.ProductName), r.Rating
                        FROM products p
                        JOIN customer_reviews r ON p.ProductID = r.ProductID
                        GROUP BY p.ProductName,r.Rating
                        ORDER BY Rating desc;"""
 
```

```python
# Result on query page vise Action
curr.execute(rating)
columns = [desc[0] for desc in curr.description]
res_avgrat = curr.fetchall() 
df7 =pd.DataFrame(res_avgrat, columns=columns)
```

```python
df7
```

```python
# Group by ProductName and take mean of Rating
avg_rating = df7.groupby('ProductName')['Rating'].mean().round(2)
```

```python
# Pivot tables For Rating
pivot1 = df7.pivot_table(
    index='ProductName',
    columns='Rating',
    aggfunc='size',   # counts occurrences
    fill_value=0
)
```

```python
# Add average rating column to the pivot table
pivot1['AverageRating'] = avg_rating
# Sort by AverageRating in descending order
pivot_avgrat = pivot1.sort_values(by='AverageRating', ascending=False)
```

```python
# Customer Retention Rate Calc
crr = """ SELECT 
                COUNT(*) * 100.0 / (SELECT COUNT(*) FROM (
                    SELECT CustomerID
                    FROM customer_journey
                    GROUP BY CustomerID
                ) AS all_customers) AS RetentionRate
            FROM (
                SELECT CustomerID
                FROM customer_journey
                GROUP BY CustomerID
                HAVING COUNT(DISTINCT VisitDate) > 1
            ) AS retained;"""
```

```python
# Result on query page vise Action
curr.execute(crr)
columns = [desc[0] for desc in curr.description]
crrr = curr.fetchall() 
df8 =pd.DataFrame(crrr, columns=columns)
```
